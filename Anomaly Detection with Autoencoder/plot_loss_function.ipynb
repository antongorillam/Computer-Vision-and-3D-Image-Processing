```python
# Plot mse loss
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

PLOT_PATH = r"/home/anton/KTH/year5/5LSH0/Anomaly Detection with Autoencoder/plots/simple_ae/"
train_loss = pd.read_csv(PLOT_PATH + "simple_ae_train_mse_loss.csv")
val_loss = pd.read_csv(PLOT_PATH + "simple_ae_val_mse_loss.csv")
train_loss["Train/Val"] = "Training loss"
val_loss["Train/Val"] = "Validation loss"
loss = pd.concat([train_loss, val_loss])
loss.rename(columns = {'Step':'Epoch'}, inplace = True)
loss.rename(columns = {'Value':'MSE Loss'}, inplace = True)
fig = sns.lineplot(data=loss, x="Epoch", y="MSE Loss", hue="Train/Val")
fig.set(title="Training vs Validation Loss Function")
plt.grid()
plt.savefig(PLOT_PATH + "training_vs_validation_loss_function")
plt.show()
plt.close()
```
